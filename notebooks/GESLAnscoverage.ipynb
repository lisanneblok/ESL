## Explore GESLA coverage

```python
import xarray as xr
```

```python
file_path = "/Users/lb962/Downloads/ESL/GESLA/data/processed/GESLA/interim/GESLA3_NorthSeaShort.nc"
# Open the dataset
GESLA_ns_dataset = xr.open_dataset(file_path)
```

```python
filtered_ds = GESLA_ns_dataset[["sea_level", "station", "latitude", "longitude"]]
```

```python
import numpy as np
# Assuming ns_data is your xarray Dataset
sea_level = filtered_ds["sea_level"]

# Set sea level values to NaN where they are below -1 or above 15
sea_level_filtered = sea_level.where((sea_level > -1) & (sea_level < 15), np.nan)
filtered_ds["q_sea_level"] = sea_level_filtered
```

# Select 1980-1982

```python
import numpy as np
filtered_ds_1980_to_1981 = filtered_ds.sel(date_time=slice('1980', '1981'))
# Get a list of unique stations
stations = filtered_ds_1980_to_1981['station'].values

# Initialize an empty list to store stations with acceptable NaN percentages
updated_stations = []

# Iterate through each station
for station in stations:
    # Calculate the percentage of NaN values for q_sea_level
    nan_percentage = filtered_ds_1980_to_1981.sel(station=station)['q_sea_level'].isnull().mean()
    
    # Check if the percentage of NaN values is less than or equal to the threshold
    if nan_percentage <= 0.08:  # Change the threshold here if needed
        updated_stations.append(station)

# Filter the dataset to include only stations with acceptable NaN percentages
filtered_NaN_1980_to_1981 = filtered_ds_1980_to_1981.sel(station=updated_stations)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create a map with a Lambert Conformal projection
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Plot ocean and land features
ax.add_feature(cfeature.LAND, edgecolor='black', facecolor='mintcream')
ax.add_feature(cfeature.OCEAN, color='lightblue')

# Plot longitude and latitude data
lon = filtered_NaN_1980_to_1981['longitude']
lat = filtered_NaN_1980_to_1981['latitude']
ax.scatter(lon, lat, s=20, c='b', marker='o', transform=ccrs.PlateCarree(), zorder=2)

# Set plot title and labels
ax.set_title('1% NaN values 1980-1981')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show plot
plt.grid(True)
plt.show()

```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

def plot_station_location(filtered_ds, station_number):
    # Extract station data based on the station number
    station_data = filtered_ds.sel(station=station_number)

    # Extract latitude and longitude values
    station_latitude = station_data['latitude'].values
    station_longitude = station_data['longitude'].values

    # Create a Cartopy PlateCarree projection
    projection = ccrs.PlateCarree()

    # Visualize the map
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(1, 1, 1, projection=projection)

    # Set extent
    ax.set_extent([-15, 10, 48, 65], crs=projection)

    # Add coastlines and borders
    ax.coastlines()
    ax.add_feature(cartopy.feature.BORDERS, linestyle=':', linewidth=1)

    # Plot the station location on the map
    ax.plot(station_longitude, station_latitude, 'o', markersize=10, color='red', transform=projection)

    # Set title and labels
    plt.title(f"Location of Station {station_number}")
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')

    # Show the plot
    plt.show()
```

```python
filtered_ds.sel(station=12).q_sea_level.plot()
plot_station_location(filtered_ds, 12)
```
